```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_df = pd.read_csv(r"./spam.csv")
print(spam_df.head())
```

```python
spam_df.groupby('Category').describe()
```

```python
spam_df['spam']=spam_df['Category'].apply(lambda x:1 if x =='spam' else 0)
```

```python
spam_df
```

```python
x_train,x_test,y_train,y_test=train_test_split(spam_df.Message,spam_df.spam,test_size=0.25)
```

```python
x_train
```

```python
cv=CountVectorizer()
x_vectorized=cv.fit_transform(spam_df['Message'])
```

```python
model=MultinomialNB()
model.fit(x_vectorized,spam_df['spam'])
```

```python
x_test_count=cv.transform(x_test)
model.score(x_test_count,y_test)
```

```python
def predict_spam(email):
    email_vectorized=cv.transform([email])
    prediction=model.predict(email_vectorized)
    return prediction[0]
```

```python
while True:
    user_input=input(" \n Enter a new mail:")
    if user_input.lower()=='exit':
        break

    prediction=predict_spam(user_input)

    if prediction=='1':
        print("this email is classified as spam.")

    else:
        print(" \n This email is not spam")
```
